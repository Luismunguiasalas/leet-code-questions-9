```python
# 1496. Path Crossing
path="NESWW"
path = "NES"
def isPathCrossing(path):
    x=0
    y=0
    s=[(x,y)]
    for n in path:
        if n=="N":y+=1
        elif n=="S":y-=1
        elif n=="E":x+=1
        elif n=="W":x-=1
        if (x,y) not in s:s.append((x,y))
        else:return False
    return True
isPathCrossing(path)
```

```python
# 100. Same Tree
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        if p is None and q is None:return True
        elif p is None and q is not None:return False
        elif p is not None and q is None:return False
        else:
            stackp=[p]
            stackq=[q]
            while stackp !=[] and stackq !=[]:
                currentp= stackp.pop()
                currentq= stackq.pop()
                if currentp.val != currentq.val:return False
                if (currentp.left and not(currentq.left))or(currentp.right and not(currentq.right)): return False
                if (currentq.left and not(currentp.left))or(currentq.right and not(currentp.right)): return False
                if currentp.left and currentq.left:
                    stackp.append(currentp.left)
                    stackq.append(currentq.left)
                if currentp.right and currentq.right:
                    stackp.append(currentp.right)
                    stackq.append(currentq.right)
            return True
```

```python
# 217. Contains Duplicate
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        l=0
        r=len(nums)-1
        d=dict()
        while l < r:
            if nums[l] not in d:
                d[nums[l]]=0
            if nums[r] not in d:
                d[nums[r]]=0
            d[nums[l]]+=1
            d[nums[r]]+=1
            if d[nums[l]]>1 or d[nums[r]]>1:return True
            l+=1
            r-=1
        return False
```

```python
# 1137. N-th Tribonacci Number
class Solution:
    def tribonacci(self, n: int) -> int:
        if n ==0:return 0
        elif n ==1:return 1
        elif n==2:return 1
        t1,t2,t3=0,1,1
        while n-2 > 0:
            t1,t2,t3=t2,t3,(t1+t2+t3)
            n-=1
        return t3
```

```python
# 448. Find All Numbers Disappeared in an Array
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        l=len(nums)
        nums=set(nums)
        a=[]
        for i in range(1,l+1):
            if i not in nums:a.append(i)
        return a
```
